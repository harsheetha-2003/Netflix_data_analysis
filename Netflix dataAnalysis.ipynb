## Principle component analysis 

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns
df_numerical = df[numerical_cols]

# Standardize the data
scaler = StandardScaler()
df_standardized = scaler.fit_transform(df_numerical)

# Perform PCA
pca = PCA(n_components=2)  # Adjust the number of components as needed
principal_components = pca.fit_transform(df_standardized)

# Create a DataFrame with the principal components
pc_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])

# Visualize the results
plt.figure(figsize=(10, 6))
sns.scatterplot(x='PC1', y='PC2', data=pc_df)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Netflix Data')
plt.show()

# Explained variance
explained_variance = pca.explained_variance_ratio_
print(f'Explained variance by component: {explained_variance}')
```

```python
# For visualization, create a dummy categorical variable
# Here we'll alternate colors for the first 100 points as an example
pc_df['Category'] = ['Category 1' if i < len(pc_df) / 2 else 'Category 2' for i in range(len(pc_df))]

# Visualize the results with different colors
plt.figure(figsize=(10, 6))
sns.scatterplot(x='PC1', y='PC2', hue='Category', palette='viridis', data=pc_df)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Netflix Data')
plt.show()

# Explained variance
explained_variance = pca.explained_variance_ratio_
print(f'Explained variance by component: {explained_variance}')
```

###  The explained variance ratio you printed earlier (not shown in the plot) tells you how much of the total variance in the data is captured by PC1 and PC2. Typically, you want these to be as high as possible for the components you plot.

### if  distinct clusters or separation are seen  along the principal components, it indicates that the data can be divided into distinct groups based on the features you used. However, in plot, the points seem to be fairly evenly spread without distinct clusters, suggesting that there may not be strong natural groupings in the data based on the first two principal components alone.

## LETS INCREASE THE PCA COMPONENTS

```python
pca = PCA(n_components=3)  # Increase to more components
principal_components = pca.fit_transform(df_standardized)

# Create a DataFrame with the principal components
pc_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2', 'PC3'])

# For visualization, we can still use a dummy categorical variable or a real one if available
pc_df['Category'] = ['Category 1' if i < len(pc_df) / 2 else 'Category 2' for i in range(len(pc_df))]

# 2D plot of PC1 vs PC3
plt.figure(figsize=(10, 6))
sns.scatterplot(x='PC1', y='PC3', hue='Category', palette='viridis', data=pc_df)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 3')
plt.title('PCA of Netflix Data (PC1 vs PC3)')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(pc_df['PC1'], pc_df['PC2'], pc_df['PC3'], c=pc_df['Category'].astype('category').cat.codes, cmap='viridis')
ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_zlabel('Principal Component 3')
plt.title('3D PCA of Netflix Data')
plt.show()
```

### Still no distinct groups appeared
